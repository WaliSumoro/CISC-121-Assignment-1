Understanding the problem
- The goal is to return a list of strings from 1 to n. so the input is integer and the out put is a list of strings. and also where multiplies of 3 we will replace it by "Fizz", where multiples of 5 we will replace it by "Buzz" or where multiples of 15 we will replace it by "FizzBuzz". if not we will leave the number itself.
- So how we can solve this is by inputting an integer, and  we will set our empty list and we will start the loop and we will use if statements inside the for loop, and finally we return the list of string.

```python
def fizz_buzz(n):
  result = []
  for i in range(1, n+1):
    if i % 15 == 0:
      result.append("FizzBuzz")
    elif i % 3 == 0:
      result.append("Fizz")
    elif i % 5 == 0:
      result.append("Buzz")
    else:
      result.append(str[i])
  return result
```

```python
def fizz_buzz(n):
    if n <= 0:
        return []
    result = []
    for i in range(1, n + 1):
        if i % 15 == 0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result
print(fizz_buzz(0))   # []
print(fizz_buzz(1))   # ['1']
print(fizz_buzz(5))   # ['1','2','Fizz','4','Buzz']
print(fizz_buzz(16))  # ... ends with 'FizzBuzz','16'
```
