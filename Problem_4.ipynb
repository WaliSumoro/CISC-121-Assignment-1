Here is also the refined code by ChatGPT

```python
def reverse_int(x: int) -> int:
    """
    Reverse a signed 32-bit integer using recursion.
    Return 0 if the reversed value overflows 32-bit signed range.
    """
    sign = 1 if x >= 0 else -1
    x = abs(x)
    LIMIT = 1 << 31  # 2**31

    def helper(n: int, acc: int) -> int:
        if n == 0:
            return acc
        return helper(n // 10, acc * 10 + (n % 10))

    r = sign * helper(x, 0)
    if r < -LIMIT or r > LIMIT - 1:
        return 0
    return r

# quick checks
assert reverse_int(123) == 321
assert reverse_int(-120) == -21
assert reverse_int(0) == 0
assert reverse_int(1534236469) == 0  # overflow -> 0
```
