```python
import re

def convert_date_to_binary(s: str) -> str:
    """
    Convert 'YYYY-MM-DD' into 'binaryYear-binaryMonth-binaryDay' without leading zeros.
    Example: '2019-01-09' -> '11111100011-1-1001'
    """
    if not re.fullmatch(r"\d{4}-\d{2}-\d{2}", s):
        raise ValueError("invalid date format, expected YYYY-MM-DD")

    y_str, m_str, d_str = s.split("-")
    y = int(y_str); m = int(m_str); d = int(d_str)

    return f"{bin(y)[2:]}-{bin(m)[2:]}-{bin(d)[2:]}"

# quick checks
assert convert_date_to_binary("2019-01-09") == "11111100011-1-1001"
assert convert_date_to_binary("0001-01-01") == "1-1-1"
assert convert_date_to_binary("2024-12-31") == f"{bin(2024)[2:]}-{bin(12)[2:]}-{bin(31)[2:]}"
# [AI Disclosure] I used ChatGPT to confirm the split→int→bin approach; I wrote and tested the final code myself.
```
