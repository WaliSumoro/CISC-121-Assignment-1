Here is also the refined code by ChatGPT

```python
def hamming_weight(n: int) -> int:
    """
    Return the number of set bits in a 32-bit unsigned integer.
    Uses n &= (n - 1) to clear the lowest set bit each step.
    Time: O(k) where k = number of 1s. Space: O(1).
    """
    n &= 0xFFFFFFFF
    cnt = 0
    while n:
        n &= (n - 1)
        cnt += 1
    return cnt

# quick checks
assert hamming_weight(0) == 0
assert hamming_weight(1) == 1
assert hamming_weight(5) == 2         # 0b0101
assert hamming_weight(0xFFFFFFFF) == 32
assert hamming_weight(-3) == 31        # as unsigned 32-bit
# [AI Disclosure] I used ChatGPT to confirm the (n & (n-1)) trick and 32-bit masking; I wrote and tested the final code myself.
```
